```python
numeric = data.select_dtypes(include = "number").columns.to_list()
print(numeric)
```

```python
values = []
for col in numeric:
    a = data[col].agg({
        'mean_value': lambda x: x.mean(),
        'min_value':  lambda x: x.min(),
        'max_value': lambda x: x.max(),
        'std.dev': lambda x: x.std(), 
        'Median': lambda x: x.median()
        })
    values.append(a)
values = pd.DataFrame(values)
values

values.to_csv('descr_stats_thesis.csv')
```

```python
corr_vars = ['price_numeric', 'age', 'years_selling', 'Height', 'Width']

corr = data[corr_vars].corr()

plt.figure(figsize=(6, 6))
sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    annot_kws={"size": 10, "weight": "bold"}
)

plt.title("Correlation Matrix", fontsize=20, fontweight='bold')
plt.xticks(fontsize=11, fontweight='bold', rotation=45)
plt.yticks(fontsize=11, fontweight='bold')

plt.savefig("corr_matrix.png", dpi=300, bbox_inches='tight')
plt.show()
```
